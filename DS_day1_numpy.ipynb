Element-Wise (Multiplication Élementaire)

La multiplication élémentaire consiste à multiplier les éléments correspondants de deux tableaux (ou matrices) de mêmes dimensions. Dans NumPy, les opérateurs arithmétiques sur les tableaux s'appliquent par éléments. Un nouveau tableau est créé et rempli avec le résultat.

```python
### Prix des produits (en euros) et Quantites
prix = np.array([10, 15, 20, 5])
quantités_vendues = np.array([100, 200, 150, 300])

print(quantités_vendues - prix)
print(prix ** 2)
print(10 * np.sin(prix))
print(quantités_vendues > 150)
```

Certaines opérations, telles que $+=$ et $*=$, permettent de modifier un tableau existant plutôt que d'en créer un nouveau.

```python
prix += 5
prix
```

```python
quantités_vendues *= 10 #### black friday
quantités_vendues
```

Matrix Multiplication (Multiplication Matricielle)

La multiplication matricielle est une opération qui combine deux matrices selon une règle spécifique. Contrairement à la multiplication élémentaire, elle nécessite que le nombre de colonnes de la première matrice corresponde au nombre de lignes de la seconde matrice. Le résultat est une nouvelle matrice.

```python
#### Matrice des ventes (lignes : commerciaux, colonnes : produits)
ventes = np.array([
    [10, 20, 15],
    [5, 10, 10],
    [8, 15, 12]
])

#### Vecteur des prix des produits
prix_produits = np.array([10, 15, 20])  # Prix par produit
revenus_commerciaux = np.dot(ventes, prix_produits)

print("Revenus totaux par commercial :", revenus_commerciaux)
print("Revenus Mode 2 : ", ventes @ prix_produits)
```

##### Travailler avec des formules mathématiques dans NumPy

Une des grandes forces de NumPy est sa capacité à traduire des formules mathématiques directement en opérations sur des tableaux, ce qui simplifie le calcul scientifique et l’implémentation de modèles. Les formules mathématiques peuvent être implémentées de manière concise et efficace grâce à l’utilisation des fonctions intégrées de NumPy.

La manière dont nous percevons les équations dans notre cerveau est suivie dans la logique d'écriture de NumPy

Par exemple, 

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} \left( Y_{\text{prédiction}_i} - Y_i \right)^2
$$


```python
#### L’Erreur quadratique moyenne
predictions = np.array([3.2, 2.8, 4.1, 3.9])
labels = np.array([3.0, 3.0, 4.0, 4.0])  

n = len(predictions)
mse = (1 / n) * np.sum(np.square(predictions - labels))
mse
```

Standard Deviation (Écart type)

$$
STD = \sqrt{\frac{1}{n} \sum_{i=1}^{n} \left( X_i - \mu \right)^2}
$$


```python
data = np.array([10, 12, 23, 23, 16, 23, 21, 16])
std_dev = np.sqrt(np.mean(np.square(data - np.mean(data))))
print("Standard Deviation: {}".format(std_dev))
```

Cross-Entropy Loss (Entropie croisée)

$$
H(p, q) = -\frac{1}{n} \sum_{i=1}^{n} \left( y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right)
$$

```python
predictions_ce = np.array([0.9, 0.8, 0.1, 0.4])  # Predicted probabilities
labels_ce = np.array([1, 1, 0, 0])               # True labels (binary)

cross_entropy = -np.mean(labels_ce * np.log(predictions_ce) + (1 - labels_ce) * np.log(1 - predictions_ce))
print("Cross-Entropy Loss:", cross_entropy)
```

#### Data Visualization - matplotlib

La visualisation des données est essentielle car elle permet de transformer des informations complexes et abstraites en représentations visuelles claires et compréhensibles. Elle facilite l’identification de tendances, de relations et d’anomalies dans de grands ensembles de données, aidant ainsi à la prise de décision éclairée.

**Matplotlib** est une bibliothèque Python puissante pour la création de visualisations statiques, animées et interactives. Elle est souvent utilisée pour générer des graphiques 2D comme des graphiques linéaires, des histogrammes, des nuages de points et bien plus. 

```python
import matplotlib.pyplot as plt

#### Génération de données
x = np.linspace(-10, 10, 100)  # 100 points entre -10 et 10
y = x**2

#### Tracé du graphique
plt.plot(x, y, label='y = x²')
plt.title("Graphique Linéaire : y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

Histogram -  Un histogramme est un graphique qui représente la distribution des valeurs d'une variable numérique sous la forme d'une série de barres.

```python
data = np.random.normal(loc=0, scale=1, size=1000)  ### Moyenne=0, Écart-type=1

# Tracé de l'histogramme
plt.hist(data, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title("Histogramme de données aléatoires")
plt.xlabel("Valeurs")
plt.ylabel("Fréquence")
plt.grid()
plt.show()
```

##### Concepts Importants en Matplotlib

1) Figure : Une figure représente la "toile" globale où tous les graphiques sont dessinés. Vous pouvez imaginer cela comme une feuille blanche.
2) Axes : Les axes sont les régions de la figure où un graphique spécifique est dessiné. Une figure peut contenir plusieurs axes.
3) Subplots : Les subplots permettent de diviser la figure en plusieurs parties pour afficher plusieurs graphiques côte à côte ou en grille.
4) Légendes, Titre, etc. : Ces éléments permettent de personnaliser et d'annoter les graphiques pour les rendre plus informatifs.

```python
#### Génération des données
values1 = np.random.randint(1, 10, size=5)
values2 = np.random.randint(1, 10, size=5)
values3 = np.random.randint(1, 10, size=5)
values4 = np.random.randint(1, 10, size=5)

labels = ['Democrats', 'Center', 'Greens', 'Socialists', 'Conservatives']

### Création de la figure et des sous-graphiques
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  ### Grille 2x2 de graphiques

### Premier graphique
axes[0, 0].pie(values1, labels=labels, autopct='%1.1f%%', startangle=90)
axes[0, 0].set_title("Élections 1")

### Deuxième graphique
axes[0, 1].pie(values2, labels=labels, autopct='%1.1f%%', startangle=90)
axes[0, 1].set_title("Élections 2")

### Troisième graphique
axes[1, 0].pie(values3, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title("Élections 3")

### Quatrième graphique
axes[1, 1].pie(values4, labels=labels, autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title("Élections 4")

### Affichage
plt.show()
```

##### Manipulation d'Images

La manipulation d'images consiste à traiter des images sous forme de tableaux numériques (matrices de pixels) afin d'extraire des informations, les transformer ou les analyser. Une image peut être représentée comme une matrice 2D (pour les images en niveaux de gris) ou 3D (pour les images en couleurs avec des canaux RGB).

Avec NumPy, nous pouvons manipuler directement les pixels d'une image, et avec Matplotlib, nous pouvons visualiser les résultats.

```python
image = plt.imread("./data/white_house.jpg")  #### Charge l'image en tant que tableau NumPy
print("Dimensions de l'image :", image.shape)

plt.imshow(image)
plt.title("Image originale")
plt.axis("off")
plt.show()
```

```python
#### Conversion en grayscale (moyenne des canaux RGB)
if len(image.shape) == 3:  ### Vérifier si l'image est en couleur
    grayscale_image = np.mean(image, axis=2)
else:
    grayscale_image = image

plt.imshow(grayscale_image, cmap="gray")
plt.title("Image en Grayscale")
plt.axis("off")
plt.show()
```

```python
negative_image = 255 - grayscale_image  ### Inverse les niveaux de gris
plt.imshow(negative_image, cmap="gray")
plt.title("Image négative")
plt.axis("off")
plt.show()
```

```python
### Réduction de la luminosité 
darker_image = grayscale_image * 0.1
plt.imshow(darker_image, cmap="gray")
plt.title("Image avec luminosité réduite")
plt.axis("off")
plt.show()
```
