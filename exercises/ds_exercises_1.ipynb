```python
#### Import numpy et matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
#### Exercice 1: Création d'un tableau 2D avec des valeurs aléatoires entre 5 et 10
array_2d = np.random.uniform(5, 10, (8, 6))  # Un tableau de 8 lignes et 6 colonnes
shape = array_2d.shape  # Récupération de la forme du tableau
last_column = array_2d[:, -1]  # Extraction de la dernière colonne
selected_rows = array_2d[1:6, :]  # Sélection des lignes entre la 2ème et la 6ème (indexation Python)
multiplied_array = array_2d * 5  # Multiplication de toutes les valeurs par 5
```

```python
#### Exercice 2: Calcul de la fonction Softmax
def compute_softmax(x):
    #### Application de la transformation Softmax
    exp_x = np.exp(x - np.max(x))  # Soustraction du max pour la stabilité numérique
    return exp_x / np.sum(exp_x)  # Normalisation pour obtenir des probabilités

random_1d_array = np.random.uniform(-10, 10, 10)  # Génération d'un tableau 1D aléatoire
softmax_scores = compute_softmax(random_1d_array)  # Calcul des scores Softmax
```

```python
#### Exercice 3: Analyse des précipitations annuelles
rainfall_data = np.random.uniform(50, 200, (5, 12))  # Génération de précipitations sur 5 ans et 12 mois
avg_rainfall_per_year = np.mean(rainfall_data, axis=1)  # Moyenne annuelle des précipitations
year_with_max_rainfall = np.argmax(avg_rainfall_per_year) + 1  # Année avec les précipitations les plus élevées
month_with_min_rainfall = np.argmin(np.min(rainfall_data, axis=0)) + 1  # Mois avec les précipitations les plus faibles

#### Visualisation des précipitations totales par an
plt.figure()
plt.bar(range(1, 6), np.sum(rainfall_data, axis=1), color = 'teal')  # Somme des précipitations par an
plt.xlabel('Année')
plt.ylabel('Précipitations Totales (mm)')
plt.title('Précipitations Totales par Année')
plt.show()
```

```python
#### Exercice 4: Manipulation d'images
image = np.random.rand(100, 100)  # Création d'une image 2D avec des valeurs aléatoires
normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))  # Normalisation entre 0 et 1
binary_image = (normalized_image > 0.5).astype(int)  # Application d'un seuil pour obtenir une image binaire

#### Affichage des images
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(normalized_image, cmap='gray')
axes[0].set_title('Image Normalisée')
axes[0].axis('off')

axes[1].imshow(binary_image, cmap='gray')
axes[1].set_title('Image Binaire')
axes[1].axis('off')

plt.show()
```

```python
#### Exercice 5: Analyse des cas de COVID-19
daily_cases = np.random.randint(0, 10000, 180)  # Génération des cas quotidiens sur 6 mois

#### Calcul des moyennes hebdomadaires
num_weeks = 180 // 7  # Nombre de semaines complètes
trimmed_cases = daily_cases[:num_weeks * 7].reshape(num_weeks, 7)  # Reshape pour diviser en semaines
weekly_averages = np.mean(trimmed_cases, axis=1)  # Calcul de la moyenne hebdomadaire

#### Affichage du graphique des cas quotidiens
# Tracer les moyennes hebdomadaires des cas COVID-19
plt.figure()
plt.bar(np.arange(1, num_weeks + 1), weekly_averages, color = 'teal')  # Tracer les moyennes hebdomadaires
plt.xlabel('Semaine')
plt.ylabel('Moyenne des Cas Hebdomadaires')
plt.title('Moyennes Hebdomadaires des Cas COVID-19 sur 6 Mois')
plt.show()
```
